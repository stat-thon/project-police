# npa 데이터

```python
datanpa = pd.read_csv('../data-police/NPA2020.csv', encoding = 'cp949')
print(datanpa.shape)

df_npa = datanpa.loc[(datanpa['EVT_CL_CD'].isin((215, 401))) & (datanpa['NPA_CL'].isin((13, 19, 31)))]
print(df_npa.shape)
# 117만건 -> 10만건
```

```python
# datetime 전처리
df_npa['Date'] = pd.to_datetime(df_npa['RECV_CPLT_DT'], format = '%Y%m%d')
df_npa['Time'] = df_npa['RECV_CPLT_TM'].astype(str)
df_npa['Time'] = df_npa['Time'].str.zfill(6)
df_npa['DateTime'] = df_npa['Date'].astype(str)
df_npa['DateTime'] = df_npa['DateTime'] + ' ' + df_npa.Time
df_npa['DateTime'] = pd.to_datetime(df_npa['DateTime'])
```

```python
df_npa.drop(['RECV_CPLT_DT', 'RECV_CPLT_TM', 'Date', 'Time'], axis = 1, inplace = True)
```
